```python
import pandas as pd

# Assuming you have already loaded the datasets

# 1. Identify Missing Values in df_customers
print("--- Missing Values in df_customers ---")
print(df_customers.isnull().sum())

# 2. Identify Missing Values in df_orders
print("\n--- Missing Values in df_orders ---")
print(df_orders.isnull().sum())

# 3. Initial Thoughts on Merging (Rationale)
print("\n--- Rationale for Merging ---")
print("We will start by merging the 'olist_orders_dataset' with other datasets to enrich the order information. Merging with 'olist_customers_dataset' using 'customer_id' will allow us to incorporate customer demographics and location data into our analysis of orders. Subsequently, we can merge with 'olist_order_items_dataset' using 'order_id' to include details about the products purchased in each order.")

# 4. Merge df_orders with df_customers
df_merged_customer_order = pd.merge(df_orders, df_customers, on='customer_id', how='left')

print("\n--- Merged Orders and Customers Data (First 5 Rows) ---")
print(df_merged_customer_order.head())
print(f"\nShape of merged df_merged_customer_order: {df_merged_customer_order.shape}")
print("\n--- Missing Values in Merged DataFrame ---")
print(df_merged_customer_order.isnull().sum())
```

```python
# Merge with order items data
df_merged_all = pd.merge(df_merged_customer_order, df_order_items, on='order_id', how='left')

print("\n--- Merged Data with Order Items (First 5 Rows) ---")
print(df_merged_all.head())
print(f"\nShape of df_merged_all: {df_merged_all.shape}")
print("\n--- Missing Values in Merged Data (with Order Items) ---")
print(df_merged_all.isnull().sum())
```

```python
# Merge with product data
df_merged_product = pd.merge(df_merged_all, df_products, on='product_id', how='left')

print("\n--- Merged Data with Product Information (First 5 Rows) ---")
print(df_merged_product.head())
print(f"\nShape of df_merged_product: {df_merged_product.shape}")
print("\n--- Missing Values in Merged Data (with Product Information) ---")
print(df_merged_product.isnull().sum())

# Merge with product category name translation
df_merged_final = pd.merge(df_merged_product, df_product_category_name_translation, on='product_category_name', how='left')

print("\n--- Final Merged Data with Product Category Name Translation (First 5 Rows) ---")
print(df_merged_final.head())
print(f"\nShape of df_merged_final: {df_merged_final.shape}")
print("\n--- Missing Values in Final Merged Data ---")
print(df_merged_final.isnull().sum())
```

```python
# Remove rows with missing order_item_id
df_cleaned = df_merged_final.dropna(subset=['order_item_id'])

print("\n--- Cleaned DataFrame (Rows with Missing order_item_id Removed) ---")
print(f"\nShape of df_cleaned: {df_cleaned.shape}")
print("\n--- Missing Values in Cleaned DataFrame ---")
print(df_cleaned.isnull().sum())
```

```python
# Remove rows with missing product_category_name
df_cleaned = df_cleaned.dropna(subset=['product_category_name'])

# Remove rows with missing product weight
df_cleaned = df_cleaned.dropna(subset=['product_weight_g'])

# Fill missing English category names with Portuguese names
df_cleaned['product_category_name_english'].fillna(df_cleaned['product_category_name'], inplace=True)

print("\n--- Final Cleaned DataFrame (Product Missing Values Handled) ---")
print(f"\nShape of df_cleaned: {df_cleaned.shape}")
print("\n--- Missing Values in Final Cleaned DataFrame ---")
print(df_cleaned.isnull().sum())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Assuming you have the 'df_cleaned' DataFrame

# 1. Feature Engineering for Delivery Time (as provided before)
df_delivery = df_cleaned.copy()

# Convert timestamp columns to datetime
timestamp_cols = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']
for col in timestamp_cols:
    df_delivery[col] = pd.to_datetime(df_delivery[col])

# Calculate actual delivery time (in days)
df_delivery['actual_delivery_time'] = (df_delivery['order_delivered_customer_date'] - df_delivery['order_purchase_timestamp']).dt.days

# Handle cases where delivery date is missing (e.g., canceled orders) - we'll remove these for now for simplicity in regression
df_delivery.dropna(subset=['actual_delivery_time'], inplace=True)

# Calculate estimated delivery lead time (in days)
df_delivery['estimated_delivery_lead_time'] = (df_delivery['order_estimated_delivery_date'] - df_delivery['order_purchase_timestamp']).dt.days

# Feature: Time to approval (in hours)
df_delivery['time_to_approval'] = (df_delivery['order_approved_at'] - df_delivery['order_purchase_timestamp']).dt.total_seconds() / 3600

# Feature: Freight value
df_delivery['freight_value'] = df_delivery['freight_value'].astype(float)

# Select relevant features for delivery time prediction
features_delivery = ['estimated_delivery_lead_time', 'time_to_approval', 'price', 'freight_value',
                     'product_name_lenght', 'product_description_lenght', 'product_photos_qty',
                     'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']

# Handle potential infinite values or NaNs after feature engineering
df_delivery.replace([np.inf, -np.inf], np.nan, inplace=True)
df_delivery.dropna(subset=features_delivery, inplace=True)

X_delivery = df_delivery[features_delivery]
y_delivery = df_delivery['actual_delivery_time']

# Split data into training and testing sets
X_train_delivery, X_test_delivery, y_train_delivery, y_test_delivery = train_test_split(X_delivery, y_delivery, test_size=0.2, random_state=42)

# 2. Train Lasso Regression Model
lasso_delivery = Lasso(alpha=0.1, random_state=42) # You might need to tune 'alpha'
lasso_delivery.fit(X_train_delivery, y_train_delivery)
y_pred_lasso_delivery = lasso_delivery.predict(X_test_delivery)

# 3. Evaluate Lasso Regression Model
def evaluate_regression(y_true, y_pred, model_name):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f"--- {model_name} Regression ---")
    print(f"RMSE: {rmse:.2f}")
    print(f"MAE: {mae:.2f}")
    print(f"R-squared: {r2:.2f}")
    return {'RMSE': rmse, 'MAE': mae, 'R-squared': r2}

lasso_metrics = evaluate_regression(y_test_delivery, y_pred_lasso_delivery, "Lasso")

regression_performance_lasso = pd.DataFrame({
    'Model': ['Lasso Regression'],
    'RMSE': [lasso_metrics['RMSE']],
    'MAE': [lasso_metrics['MAE']],
    'R-squared': [lasso_metrics['R-squared']]
})

print("\n--- Lasso Regression Performance ---")
print(regression_performance_lasso)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Assuming you have the 'df_cleaned' DataFrame and have performed the feature engineering steps
# as shown in the previous Lasso Regression code.

# 1. Feature Engineering for Delivery Time (Ensure this part has been run)
df_delivery = df_cleaned.copy()

# Convert timestamp columns to datetime
timestamp_cols = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']
for col in timestamp_cols:
    df_delivery[col] = pd.to_datetime(df_delivery[col])

# Calculate actual delivery time (in days)
df_delivery['actual_delivery_time'] = (df_delivery['order_delivered_customer_date'] - df_delivery['order_purchase_timestamp']).dt.days

# Handle cases where delivery date is missing
df_delivery.dropna(subset=['actual_delivery_time'], inplace=True)

# Calculate estimated delivery lead time (in days)
df_delivery['estimated_delivery_lead_time'] = (df_delivery['order_estimated_delivery_date'] - df_delivery['order_purchase_timestamp']).dt.days

# Feature: Time to approval (in hours)
df_delivery['time_to_approval'] = (df_delivery['order_approved_at'] - df_delivery['order_purchase_timestamp']).dt.total_seconds() / 3600

# Feature: Freight value
df_delivery['freight_value'] = df_delivery['freight_value'].astype(float)

# Select relevant features
features_delivery = ['estimated_delivery_lead_time', 'time_to_approval', 'price', 'freight_value',
                     'product_name_lenght', 'product_description_lenght', 'product_photos_qty',
                     'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']

# Handle potential infinite values or NaNs
df_delivery.replace([np.inf, -np.inf], np.nan, inplace=True)
df_delivery.dropna(subset=features_delivery, inplace=True)

X_delivery = df_delivery[features_delivery]
y_delivery = df_delivery['actual_delivery_time']

# Split data
X_train_delivery, X_test_delivery, y_train_delivery, y_test_delivery = train_test_split(X_delivery, y_delivery, test_size=0.2, random_state=42)

# 2. Train Random Forest Regression Model
rf_delivery = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
rf_delivery.fit(X_train_delivery, y_train_delivery)
y_pred_rf_delivery = rf_delivery.predict(X_test_delivery)

# 3. Evaluate Random Forest Regression Model
def evaluate_regression(y_true, y_pred, model_name):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f"--- {model_name} Regression ---")
    print(f"RMSE: {rmse:.2f}")
    print(f"MAE: {mae:.2f}")
    print(f"R-squared: {r2:.2f}")
    return {'RMSE': rmse, 'MAE': mae, 'R-squared': r2}

rf_metrics = evaluate_regression(y_test_delivery, y_pred_rf_delivery, "Random Forest")

regression_performance_rf = pd.DataFrame({
    'Model': ['Random Forest Regression'],
    'RMSE': [rf_metrics['RMSE']],
    'MAE': [rf_metrics['MAE']],
    'R-squared': [rf_metrics['R-squared']]
})

print("\n--- Random Forest Regression Performance ---")
print(regression_performance_rf)
```
